# Data Load


```
file_path = "/gdrive/My Drive/colab/Support Vector Machine/week5/SMSSpamCollection"

with open(file_path,'r',encoding='utf8') as inFile:
  lines = inFile.readlines()

x_data, y_data = [], []
for line in lines:
  pieces = line.strip().split('\t')
  label, sentence = pieces[0], pieces[1].lower()
  
  x_data.append(sentence)
  y_data.append(label)
  
print("x_data의 개수 : " + str(len(x_data)))
print("y_data의 개수 : " + str(len(y_data)))
```

# Tokenizer 라이브러리를 사용하여 입력 문장을 index로 치환

```
from keras.preprocessing.text import Tokenizer
import pandas as pd

# spam, ham 라벨을 대응하는 index로 치환하기위한 딕셔너리
label2index = {'spam':0, 'ham':1}
index2label = {0:"spam", 1:"ham"}

# indexing 한 데이터를 넣을 리스트 선언
indexing_x_data, indexing_y_data = [], []

for label in y_data:
  indexing_y_data.append(label2index[label])

tokenizer = Tokenizer()

# x_data를 사용하여 딕셔너리 생성
tokenizer.fit_on_texts(x_data)                

# x_data에 있는 각 문장들을 one-hot 벡터의 합으로 치환하고 그 결과값을 indexing_x_data에 저장
indexing_x_data = tokenizer.texts_to_matrix(x_data, mode='count').tolist()

n_spam = indexing_y_data.count(0)
n_ham = indexing_y_data.count(1)

print('spam 개수:',indexing_y_data.count(0),'\nham개수:',indexing_y_data.count(1))
print(tokenizer.texts_to_matrix(x_data))
```

```
print(len(tokenizer.word_index))
print(len(tokenizer.texts_to_matrix(x_data)[0]))   
print(sum([x[0] for x in indexing_x_data]))

mod_indexing_x_data = [x[1:] for x in indexing_x_data]
print(len(mod_indexing_x_data[0]))

df_indexing = pd.DataFrame(mod_indexing_x_data, columns= tokenizer.word_index, index= ['doc'+str(x) for x in range(1,1501)])
col_sum = df_indexing.sum()
df_indexing.loc['sum'] = col_sum
df_indexing.sort_values(by='sum')

df_indexing
```

<h2><b>카이제곱 중요도를 이용한 Keyword 추출</b></h2>

```
#Spam/Ham sentence 나누기
spam_sentence, ham_sentence= [], []

for i in range(len(x_data)):
    if indexing_y_data[i] == label2index['spam']:
        spam_sentence.append(x_data[i])  
    else: ham_sentence.append(x_data[i])

print('spam 개수:',len(spam_sentence),'\nham개수:',len(ham_sentence))
print(spam_sentence)
```

```
#Document Frequency 함수 생성
def df(t,docs):
    df= 0
    for doc in docs:
        words= doc.split()
        df += t in words

    return df

#카이제곱 중요도 함수 생성
    # A: number of word in spam
    # B: number of word in ham
    # C: number of word not in spam
    # D: number of word not in ham

def chi_square(t):
    A = df(t, spam_sentence)
    B = df(t, ham_sentence)
    C = n_spam - A
    D = n_ham - B

    try: chi = ((A+B+C+D)*(A*D-C*B)**2) / ((A+C)*(B+D)*(A+B)*(C+D))
    except: chi = ((A+B+C+D)*(A*D-C*B)**2) / (1500*1500)
  
    return chi
```

```
# Word Corpus 생성
word_corpus= []
for i in x_data:
    word_corpus += i.split()

word_corpus = list(set([x for x in word_corpus if len(x)>1]))

# 카이제곱 중요도
chi_weight = []
for word in word_corpus:
    chi_weight.append(chi_square(word))

df_chi_weight = pd.DataFrame(chi_weight, columns=['chi weight'] ,index= word_corpus)
df_chi_weight = df_chi_weight.sort_values(by='chi weight',ascending=False).iloc[:200]
df_chi_weight
```

```
# chi_square matrix
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

x =vectorizer.fit(df_chi_weight.index)
matrix = x.transform(x_data).toarray()

df_chi = pd.DataFrame(matrix , columns= vectorizer.get_feature_names(), index= ['doc'+str(x) for x in range(1,1501)])
df_chi
```

```
df_indexing_300 = df_indexing.iloc[:,:300]
df_data = pd.concat([df_indexing_300,df_chi], axis=1)
df_data= df_data.iloc[:1500]

data = []
for i in range(len(df_data)):
    data.append(df_data.iloc[i].tolist())

df_data
```

<h2><b>SVM 모델 학습</b></h2>

```
from sklearn.svm import SVC

# 전체 데이터를 8:2의 비율로 나누어 학습 및 평가 데이터로 사용
number_of_train_data = int(len(indexing_x_data)*0.8)

train_x, train_y = data[:number_of_train_data], indexing_y_data[:number_of_train_data]
test_x, test_y = data[number_of_train_data:], indexing_y_data[number_of_train_data:]

print("train_x의 개수 : " + str(len(train_x)))
print("train_y의 개수 : " + str(len(train_y)))
print("test_x의 개수 : " + str(len(test_x)))
print("test_y의 개수 : " + str(len(test_y)))

svm = SVC(kernel='rbf')
svm.fit(train_x, train_y)
```

<h2><b>SVM 모델을 이용한 SPAM 여부 측정</b></h2>

```
from sklearn.metrics import accuracy_score

predict = svm.predict(test_x)

print('Accuracy: %.2f' % accuracy_score(test_y, predict))
for index in range(len(x_data[number_of_train_data:number_of_train_data+5])):
  print()
  print("문장 : ", x_data[index])
  print("정답 : ", index2label[test_y[index]])
  print("모델 출력 : ", index2label[predict[index]])
```
