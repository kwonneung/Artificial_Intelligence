```
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split
import graphviz
import os

data = load_iris()

x = data.data
y = data.target
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)

print(x_train,'\n', y_train)
print(x_test,'\n', y_test)
```

```
print(data["feature_names"])
print(data["data"])
print(data["target"])
print(data["target_names"])
```

```
#모델생성 및 모델링
clf = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth=None, random_state=0)
clf.fit(x_train,y_train)

#정확도 평가
print("Train Accuracy {:3f}".format(clf.score(x_train,y_train)))
print("Test Accuracy {:3f}".format(clf.score(x_test, y_test)))
```

```
#Visualization
dot_data = tree.export_graphviz(clf, out_file=None, filled=True, class_names=data.target_names) #뒤에 두 속성: 결과를 표시하고 색칠
graph = graphviz.Source(dot_data) 
graph
```
